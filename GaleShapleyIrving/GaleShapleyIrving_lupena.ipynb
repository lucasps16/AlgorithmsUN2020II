# Best case

```

def rotateArray(arr):
  temp = arr[0]
  for i in range(len(arr)-1):
    arr[i] = arr[i+1]
  arr[len(arr)-1] = temp
  return arr
		


def generateBestCase(n):#Funtion to generate the best case, receives the number of possible couples
  #The first n/2 are considered men, and the second n/2 are considered women
  pref1 = list(range(1,n+1))
  pref2 = list(range(n+1,(2*n)+1))
  listOfPref = [0 for i in range((2*n))] #List of lists of preferences
  for i in range(n):
    listOfPref[i]=pref2.copy()
    listOfPref[i+n]=pref1.copy()
    pref1 = rotateArray(pref1)
    pref2 = rotateArray(pref2)
  for i in range(len(listOfPref)):
    print("Person "+str(i+1)+ " list of preference: "+str(listOfPref[i])+"\n")

```

```
print("BEST CASE FOR N=4")
generateBestCase(4)
```

```
print("BEST CASE FOR N=40")
generateBestCase(40)
```

# Worst Case

```
def generateWorstCase(n):#Funtion to generate the worst case, receives the number of possible couples 
  #The first n/2 are considered men, and the second n/2 are considered women
  pref1 = list(range(1,n+1))
  pref2 = list(range(n+1,(2*n)+1))
  listOfPref = [0 for i in range((2*n))] #List of lists of preferences
  for i in range(n):
    listOfPref[i]=pref2.copy()
    listOfPref[i+n]=pref1.copy()
  for i in range(len(listOfPref)):
    print("Person "+str(i+1)+ " list of preference: "+str(listOfPref[i])+"\n")
```

```
def generateRandCase(n):#Funtion to generate a random case, receives the number of possible couples
  #The first n/2 are considered men, and the second n/2 are considered women
  pref1 = list(range(n))
  pref2 = list(range(n,(2*n)))
  listOfPref = [0 for i in range((2*n))] #List of lists of preferences
  for i in range(n):
    random.shuffle(pref1)
    random.shuffle(pref2)
    listOfPref[i]=pref2.copy()
    listOfPref[i+n]=pref1.copy()
  #for i in range(len(listOfPref)):
    #print("Person "+str(i+1)+ " list of preference: "+str(listOfPref[i])+"\n")
  return listOfPref
```

# Toward Data Science Simply explained Algorithm

```
import pandas as pd
import numpy as np
from collections import Counter
from copy import copy
```

```
man_list = ['a', 'b', 'c', 'd']
women_list = ['A', 'B', 'C', 'D']
```

```
women_df = pd.DataFrame({'A': [3,4,2,1], 'B': [3,1,4,2], 'C':[2,3,4,1], 'D':[3,2,1,4]})
women_df.index = man_list
```

```
man_df = pd.DataFrame({'A': [1,1,2,4], 'B': [2,4,1,2], 'C':[3,3,3,3], 'D':[4,2,4,1]})
man_df.index = man_list
```

```
#dict to control which women each man can make proposals
women_available = {man:women_list for man in man_list}
# waiting list of men that were able to create pair on each iteration
waiting_list = []
# dict to store created pairs
proposals = {}
# variable to count number of iterations
count = 0
```

```
while len(waiting_list)<len(man_list):
    # man makes proposals
    for man in man_list:
        if man not in waiting_list:
            # each man make proposal to the top women from it's list
            women = women_available[man]
            best_choice = man_df.loc[man][man_df.loc[man].index.isin(women)].idxmin()
            proposals[(man, best_choice)]=(man_df.loc[man][best_choice],
                                                 women_df.loc[man][best_choice])
    # if women have more than one proposals 
    # she will choose the best option
    overlays = Counter([key[1] for key in proposals.keys()])
    # cycle to choose the best options
    for women in overlays.keys():
        if overlays[women]>1:
            # pairs to drop from proposals
            pairs_to_drop = sorted({pair: proposals[pair] for pair in proposals.keys() 
                    if women in pair}.items(), 
                   key=lambda x: x[1][1]
                  )[1:]
            # if man was rejected by woman
            # there is no pint for him to make proposal 
            # second time to the same woman
            for p_to_drop in pairs_to_drop:
                del proposals[p_to_drop[0]]
                _women = copy(women_available[p_to_drop[0][0]])
                _women.remove(p_to_drop[0][1])
                women_available[p_to_drop[0][0]] = _women
    # man who successfully created pairs must be added to the waiting list 
    waiting_list = [man[0] for man in proposals.keys()]
    # update counter
    count+=1
```

# Geeks for Geeks Algorithm

```

# This function returns true if 
# woman 'w' prefers man 'm1' over man 'm' 
def wPrefersM1OverM(prefer, w, m, m1,N): 
	
	# Check if w prefers m over her 
	# current engagment m1 
	for i in range(N): 
		
		# If m1 comes before m in lisr of w, 
		# then w prefers her current engagement, 
		# don't do anything 
		if (prefer[w][i] == m1): 
			return True

		# If m cmes before m1 in w's list, 
		# then free her current engagement 
		# and engage her with m 
		if (prefer[w][i] == m): 
			return False

# Prints stable matching for N boys and N girls. 
# Boys are numbered as 0 to N-1. 
# Girls are numbereed as N to 2N-1. 
def stableMarriage(prefer,N): 

  wPartner=[-1 for i in range(N)]
  mFree = [False for i in range(N)]
  freeCount = N
  nIterations = 0
  while(freeCount>0):
    m=0
    while (m<N):
      if (mFree[m] == False):
        break
      m+=1
    i = 0
    while (i<N and mFree[m]==False):
      nIterations += 1
      w=prefer[m][i]
      if (wPartner[w-N] == -1):
        wPartner[w-N] = m
        mFree[m]=True
        freeCount -= 1
      else:
        m1=wPartner[w-N]
        if (wPrefersM1OverM(prefer,w,m,m1,N)==False):
          wPartner[w-N] = m
          mFree[m]=True
          mFree[m1]=False
        i += 1
  #print("Woman "," Man")
  #for i in range(N):
    #print(i+N, "\t", wPartner[i])
  return nIterations


# This code is contributed by Mohit Kumar 
```

# **Run the Gale-Shapley Algorithm 100 times using random permutations as preference lists for n=4 and 40. Find the minimum, maximum and average number of iterations and plot the distribution of the number of iterations.**

```
def run100(N):
  nIterations = []
  for i in range(100):
    prefList = generateRandCase(N)
    nIterations.append(stableMarriage(prefList,N))

  maxIterations = max(nIterations)
  minIterations = min(nIterations)
  avrgIterations = (sum(nIterations)/len(nIterations))

  print("Average: "+ str(avrgIterations)+"\n")
  print("Minimum: "+ str(minIterations)+" \n")
  print("Maximum: "+ str(maxIterations) +"\n")

  plt.hist(nIterations)
  plt.xlabel("Number of iterations")
  plt.show()
```
