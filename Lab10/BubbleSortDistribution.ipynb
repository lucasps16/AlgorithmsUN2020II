```
import itertools as tools
import matplotlib.pyplot as plt
```

```
def swap(v,i,j): #Swaps two adjacent numbers if they're not ordered
  temp = v[i] #Temp variable to store one of the numbers
  v[i]=v[j]
  v[j]=temp

def bubbleSort(v):
  pasos=0 #Steps to order the array
  for i in range(len(v)):
    for j in range(len(v)-1,i,-1):
      if(v[j]<v[j-1]): #Check if two adjacent numbers are ordered
        swap(v,j,j-1) #Swap the number if they're not ordered 
        pasos+=1 #add 1 to the steps
  return pasos


def stepDistribution(v):
  permutations = tools.permutations(v)
  steps = []
  for i in permutations:
    steps.append(bubbleSort(list(i)))
  plt.title("BubbleSort step distribution (array size = " + str(len(v))+")")
  plt.hist(steps,max(steps)+1,edgecolor="black")
  plt.xlabel("Steps")
  plt.show()

```

```
stepDistribution([1,2,3])
```
